```python
spam = pd.read_csv('/kaggle/input/lets-surpass-the-hosts-bayesian-model/train_df.csv')
```

```python
# all the data is provided as numbers,
# which makes it unnecessary to preprocess the data
spam.info()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler

y = spam.Target
X = spam.drop('Target', axis=1)

scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)

X_train, X_valid, y_train, y_valid = train_test_split(X_scaled, y, train_size=0.8, test_size=0.2)
```

```python
# train the model
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score

model = GaussianNB()
model.fit(X_train, y_train)

predictions = model.predict(X_valid)
accuracy = accuracy_score(predictions, y_valid)
print(f"Accuracy: {accuracy}")
```

```python
submission = pd.read_csv('/kaggle/input/lets-surpass-the-hosts-bayesian-model/submission.csv')
submission
```

```python
# get the test predictions
spam_test = pd.read_csv('/kaggle/input/lets-surpass-the-hosts-bayesian-model/test_df.csv')

X_test_scaled = scaler.transform(spam_test)

final_predictions = model.predict(X_test_scaled)

submission = pd.DataFrame({
    'id': spam_test['id'],
    'Target': final_predictions
})

submission.to_csv('submission.csv', index=False)
```
