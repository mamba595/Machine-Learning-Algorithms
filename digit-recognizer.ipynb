```python
# get a glimpse on the data
digits = pd.read_csv('/kaggle/input/digit-recognizer/train.csv')
```

```python
# as we can see here, no columns should be modified or deleted
digits.head()
```

```python
# as we can see, values go from 0 to 254
digits.describe()
```

```python
# we separate the target ('label') from the data
y = digits.label
X = digits.drop('label', axis=1)
```

```python
# normalizing the values before splitting the data
X = X / 255
```

```python
# splitting the data into training and validation sets
from sklearn.model_selection import train_test_split

X_train, X_valid, y_train, y_valid = train_test_split(X, y, train_size=0.8, test_size=0.2, random_state=0)
```

```python
# defining the model using Stochastic Gradient Descent Classifier
from sklearn.linear_model import SGDClassifier

# the loss function separates the model from logistic regression
model = SGDClassifier(loss='hinge') 
```

```python
# training the model and getting the predictions
from sklearn.metrics import hinge_loss as sklearn_hinge_loss

model.fit(X_train, y_train)
predictions = model.decision_function(X_valid)

# hinge loss function: (1/n) * sum(max(0, 1 - y_valid[i] * predictions[i]))
loss = sklearn_hinge_loss(y_valid, predictions)

print(f'Loss: {loss}')
```
